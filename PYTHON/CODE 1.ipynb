```python
from scipy.io import arff
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Load ARFF file
data = arff.loadarff(r'D:\New folder (2)\php50jXam.arff')
df = pd.DataFrame(data[0])

# Combine V1 and V2 into a single array for clustering
features = ['V2', 'V1']  # You can modify this based on your preference
data_for_clustering = df[features]

# Number of runs for KMeans
num_runs = 10

# Plot clustered data for multiple runs
for run in range(num_runs):
    # Fit KMeans model
    km_res = KMeans(n_clusters=2).fit(data_for_clustering)
    
    # Assign cluster labels to the DataFrame
    df[f'Cluster_{run+1}'] = km_res.labels_

    # Plot clustered data
    plt.figure(figsize=(8, 6))
    plt.scatter(data_for_clustering[features[0]], data_for_clustering[features[1]], c=km_res.labels_, cmap='viridis')
    plt.scatter(km_res.cluster_centers_[:, 0], km_res.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Cluster Centers')
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.title(f'KMeans Clustering - Run {run+1}')
    plt.legend()
    plt.show()

# Compare the results
cluster_columns = [f'Cluster_{run+1}' for run in range(num_runs)]
cluster_counts = df[cluster_columns].apply(lambda x: x.value_counts())

# Print the counts for each cluster in each run
print("Cluster Counts:")
print(cluster_counts)

# Check stability by calculating the pairwise Jaccard similarity between clusterings
jaccard_similarity = np.zeros((num_runs, num_runs))
for i in range(num_runs):
    for j in range(num_runs):
        if i != j:
            common_elements = np.sum(df[cluster_columns[i]] == df[cluster_columns[j]])
            jaccard_similarity[i, j] = common_elements / len(df)

# Print Jaccard similarity matrix
print("\nJaccard Similarity Matrix:")
print(pd.DataFrame(jaccard_similarity, columns=[f'Run {i+1}' for i in range(num_runs)], index=[f'Run {i+1}' for i in range(num_runs)]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the initial wave function
def initial_psi(x):
    return 3.99471 * np.exp(-((x - 0.25) / 0.05)**2)

# Define the basis function
def basis(n, x):
    return np.sqrt(2) * np.sin(n * np.pi * x)

# Numerically integrate the product of initial wave function and basis function
def coefficient(n):
    integrand = lambda x: initial_psi(x) * basis(n, x)
    result, _ = quad(integrand, 0, 1)
    return result

# Calculate coefficients for n = 1 to 30
coefficients = [coefficient(n) for n in range(1, 31)]

# Define the psi function
def psi(x, t):
    return sum(c * basis(n, x) * np.exp(-1j * n**2 * t) for n, c in enumerate(coefficients, start=1))

# Values for x and t
x_values = np.linspace(0, 1, 100)
t_values = np.linspace(0, 1, 100)

# Create a meshgrid for x and t
x_mesh, t_mesh = np.meshgrid(x_values, t_values)

# Calculate psi for each combination of x and t
psi_values = psi(x_mesh, t_mesh)

# Plot the real part of psi
plt.figure(figsize=(10, 8))
plt.imshow(np.real(psi_values), extent=(0, 1, 0, 1), origin='lower', aspect='auto', cmap='viridis')
plt.colorbar(label='Real Part of Psi')
plt.xlabel('x')
plt.ylabel('t')
plt.title('Real Part of Psi in 2D')
plt.show()
```
